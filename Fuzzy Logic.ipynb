### 1.1 Fungsi Keanggotaan Permintaan

```python
Permintaan_Rendah       = fuzz.trapmf(Permintaan, [1, 1, 500, 575])
Permintaan_Sedang         = fuzz.trimf(Permintaan, [500, 575, 650])
Permintaan_Tinggi         = fuzz.trimf(Permintaan, [575, 650, 650])
```

<div class="alert alert-warning">

<b>Himpunan fungsi keanggotaan Permintaan</b> 

 <ul>
  <li>1. Untuk himpunan Rendah digunakan fungsi trapesium (trapmf) dengan nilai 1, 1, 500, 575</li>
  <li>2. Himpunan Sedang menggunakan fungsi segitiga (trimf) dengan range nilai 500, 575, 650</li>
  <li>3. Himpunan Tinggi sama seperti himpunan Sedang yaitu menggunakan fungsi segitiga (trimf) dengan nilai 575, 650, 650</li>  
</ul>
</div>

1.1.1 Permintaan[Rendah] = ( 0 - 575 ) 
    
\begin{equation}
  Permintaan['Rendah'] = \mu PermintaanRendah(x)=\begin{cases}
    1, & \text{$ \leq 500$}.\\
    \left.(575 - x) \,  \right/(575 - 500) , & \text{$500 \leq x \leq 575$}.\\
    0, & \text{$x\geq 575$}.
  \end{cases}
  \end{equation}

<div class="alert alert-warning">

<b>Fungsi keanggotaan Permintaan rendah</b> 

 <ul>
  <li>Persamaan Permintaan Rendah. 1 jika kurang dari 500, 0 jika lebih dari 575 dan (575 - x)/575 - 500) jika 500 - 575 </li>  
</ul>
</div>

```python
box = dict(facecolor='yellow', pad=5, alpha=0.2)
grafik, (PermintaanRendah) = plt.subplots(ncols= 1 ,facecolor='white', figsize=(5,5))
PermintaanRendah.plot(Permintaan, Permintaan_Rendah, 'red', linewidth=1.5, label='Rendah')
PermintaanRendah.set_title('Permintaan Rendah',bbox=box)
PermintaanRendah.grid()
PermintaanRendah.legend()
```

<div class="alert alert-warning">

<b>Penjelasan Line Code Himpunan Permintaan Rendah</b> 

 <ul>

  <li>3. Line ke 3 digunakan untuk memberi label, warna garis dan plot Permintaan Rendah terhadap Permintan</li>

 fuzzy
diperoleh dengan cara mengambilnilai maksimum aturan, kemudian menggunakan nilai tersebut untuk memodifikasi daerah
 fuzzy
dan mengaplikasikannya ke
output
dengan menggunakan operato OR (gabungan). Jika semua proporsi telah dievaluasi, maka
output
akan berisi suatu himpunan
 fuzzy
yang merefleksikan kontribusi dari tiap-tiap proporsi
